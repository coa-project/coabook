```
# testing the correct install of pycoa
import importlib
coa_module_info = importlib.util.find_spec("coa")
if coa_module_info == None:
  raise ModuleNotFoundError("The coa module is not found. Please check your install")

import coa
if coa.__version__ != 'v1.0':
  raise ImportError("The pycoa package version is incorrect. Expecting v1.0")
```

## Importing the front module 

```
import coa.front as cf
```

## Using the `front` module to produce simple plot, map, histo or getting the data

```
cf.plot(where=['France', 'Italy', 'United kingdom'], which='deaths', what='cumul')
```

```
cf.map(where=['world'])
```

```
cf.hist(where='middle africa', which='confirmed')
```

```
cf.get(where=['usa'], what='daily', which='recovered',output='pandas')
```
